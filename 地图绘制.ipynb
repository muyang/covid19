```python
import pandas as pd
import numpy as np
from pyecharts.charts import Geo  # 导入地理信息处理模块
from pyecharts import options as opts  # 配置
from pyecharts.render.snapshot import make_snapshot
from snapshot_selenium import snapshot
from pyecharts .globals import ChartType, SymbolType, GeoType
```

```python
data=pd.read_csv("data/time_series_19-covid-Confirmed_archived_0325.csv")
data['Province/State']=data['Province/State'].fillna(0)
data.head()
```

```python
# 人数求和
geo_sum=[data.iloc[i,4:].sum() for i in range(len(data))]
geo_sum
```

```python
geo_sum=pd.Series(data=geo_sum)
geo_sum
```

```python
# 筛选地图数据
data_geo=data.loc[:,['Province/State','Country/Region','Lat','Long']]
data_geo['sum']=geo_sum
data_geo.head()
```

```python
geo_sight_coord = {data_geo['Country/Region'][i] if data_geo['Province/State'][i]==0 else data_geo['Province/State'][i]+','+data_geo['Country/Region'][i]: [data_geo['Long'][i],data_geo['Lat'][i],data_geo['sum'][i]]  for i in range(len(data_geo))}  # 构造位置字典数据
geo_sight_coord
```

```python
# 绘制地图
g = Geo(init_opts=opts.InitOpts(width="1800px", height="1000px",))
g.add_schema(maptype="world",zoom=1.2,itemstyle_opts=opts.ItemStyleOpts(color='#333333',border_color='#FFFF22'))
list1 = []
for i in geo_sight_coord:  # 对地理点循环
    g.add_coordinate(str(i), geo_sight_coord[i][0], geo_sight_coord[i][1])  # 追加点位置
    list1.append([i, geo_sight_coord[i][2]])
g.add("", list1,symbol_size=10,label_opts=opts.LabelOpts(
            position="top",
            is_show=True,
            formatter='{b}'))
g.set_series_opts(label_opts=opts.LabelOpts(is_show=False,formatter='{b}'))
g.set_global_opts(title_opts=opts.TitleOpts(title="全球地图",pos_left='900px',title_textstyle_opts=opts.TextStyleOpts(font_size=30)))
g.render("data/全球地图.html")
```
