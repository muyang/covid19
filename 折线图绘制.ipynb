```python
import pandas as pd
import numpy as np
from pyecharts.charts import Line
from pyecharts import options as opts 
from pyecharts.render.snapshot import make_snapshot
from snapshot_selenium import snapshot
```

```python
data=pd.read_csv('data/time_series_19-covid-Confirmed_archived_0325.csv')
data.head()
```

```python
list(set(data["Country/Region"].values.tolist()))
```

# G20国家数据(19个国家)

```python
countrys20=["China","Argentina","Australia","Brazil","Canada", "France", "Germany", 
            "India", "Indonesia", "Italy", "Japan", "Korea, South", "Mexico",
            "Russia","Saudi Arabia", "South Africa", "Turkey","United Kingdom","US"]
```

```python
index=[True if data['Country/Region'][i] in countrys20 else False for i in range(len(data))]
data=data[index]
data.head()
```

```python
data_new=data[[x for x in data.columns if x not in ['Province/State','Lat','Long']]]
data_new.head()
```

```python
data_new=data_new.groupby(by=['Country/Region']).sum().reset_index()
data_new.head()
```

```python
# 7日均值数据
data_average7=data_new.copy()
data_average7.head()
```

```python
for i in range(4,len(data_new.columns)-3):
        average=data_new.iloc[:,i-3]
        for j in range(i-2,i+4):
            average=data_new.iloc[:,j]+average
        data_average7.iloc[:,i]=average//7
data_average7
```

```python
# 累加人数数据
data_sum=data_new.copy()
for i in range(2,len(data_sum.columns)):
    data_sum.iloc[:,i]=data_sum.iloc[:,i]+data_sum.iloc[:,i-1]
data_sum
```

```python
# 绘图函数
def func(data,title):
    x=data_new.columns.to_list()[1:]
    line =Line(init_opts=opts.InitOpts(width="1600px", height="900px",))
    line.add_xaxis(xaxis_data=x)
    for i in range(len(data)):
        country=list(data.iloc[i]).pop(0)
        y=[str(i) for i in list(data.iloc[i])[1:]]
        line.add_yaxis(series_name=country, y_axis=y,symbol='circle',is_smooth = True)
    line.set_global_opts(title_opts=opts.TitleOpts(title=title),
                         legend_opts = opts.LegendOpts(type_="scroll", pos_right="right", orient="vertical"),
                         xaxis_opts=opts.AxisOpts(name='日期'),
                         yaxis_opts=opts.AxisOpts(name='人数')
                        )
    line.set_series_opts(
        label_opts=opts.LabelOpts(is_show=False)
    )
    line.render('data/'+title+'.html')
    
```

```python
func(data_new,'各国折线图')
```

```python
func(data_sum,'各国累加折线图')
```

```python
func(data_average7,'各国平均折线图')
```
